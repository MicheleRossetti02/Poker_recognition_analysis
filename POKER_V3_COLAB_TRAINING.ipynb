## 📥 Step 2: Download Dataset V3 da Roboflow

```python
from roboflow import Roboflow
import os

# Autenticazione Roboflow
rf = Roboflow(api_key="CUhu9vmliKIKhDxLNMVU")
project = rf.workspace("pokergtobot").project("poker-gto-production")

# Download Dataset V3
print("📥 Downloading Dataset V3 (3,733 images)...")
dataset = project.version(3).download("yolov8")

print(f"✅ Dataset scaricato in: {dataset.location}")
print(f"📊 Data YAML: {dataset.location}/data.yaml")
```

## 🚀 Step 3: Training YOLOv11 (Full Quality - 100 Epochs)

```python
from ultralytics import YOLO

# Inizializza modello YOLOv11n
model = YOLO('yolo11n.pt')

# Training con configurazione MAX QUALITY
print("🔥 Avvio Training YOLOv11...")
results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=100,
    imgsz=640,          # Full resolution
    batch=16,           # Batch size (la GPU lo regge)
    device=0,           # GPU
    plots=True,
    save=True,
    save_period=5,      # Salva ogni 5 epoche
    project='poker_v3_colab',
    name='train_full_quality',
    patience=50,        # Early stopping se non migliora per 50 epoche
    verbose=True
)

print("\n✅ Training completato!")
print(f"📁 Weights salvati in: {results.save_dir}")
```

## 📊 Step 4: Validazione Modello

```python
# Valida il modello sul validation set
metrics = model.val()

print("\n📊 Metriche Finali:")
print(f"mAP50: {metrics.box.map50:.4f}")
print(f"mAP50-95: {metrics.box.map:.4f}")
print(f"Precision: {metrics.box.mp:.4f}")
print(f"Recall: {metrics.box.mr:.4f}")
```

## 💾 Step 5: Export Weights per Download

```python
import shutil
from google.colab import files

# Path del best.pt
weights_path = f"{results.save_dir}/weights/best.pt"

# Copia in root per facilità di download
shutil.copy2(weights_path, "poker_v3_best.pt")

print("📥 Download del file best.pt...")
print("⚠️ NOTA: Il download potrebbe richiedere alcuni secondi.")
print("\nDopo il download, sposta il file in:")
print("POKER_GTO_BOT_V3/weights/best.pt")
print("\nPremendo il pulsante qui sotto il file verrà scaricato sul tuo Mac.")

# Download automatico
files.download('poker_v3_best.pt')
```
